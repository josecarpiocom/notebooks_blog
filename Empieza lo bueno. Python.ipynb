### Primer proyecto Python. Análisis de acciones.

Vamos a utilizar pytho  para descargarnos datos de acciones a través de la API de Google y graficarlos.

Importamos las librerías necesarias.

```python
%matplotlib inline  
import pandas
import pandas_datareader.data as web
import datetime
```

Definimos fechas de inicio y fin para la descarga de histórico.

```python
inicial = datetime.datetime(2014, 1, 1)
final = datetime.datetime.today()
tickers = ['AAPL', 'AMZN'] # apple y amazon
panel_data = web.DataReader(tickers, 'google', inicial, final)
```

Creamos dataframe con los precios de cierre y graficamos

```python
close = panel_data.ix['Close']
#normalizamos para que empiezen en el mismo nivel
close = close/close.ix[0]
close.plot()
```
